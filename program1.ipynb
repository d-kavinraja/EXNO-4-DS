# FEATURE SELECTION

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
data=pd.read_csv("income.csv",na_values=[' ?'])
data
```

```python
data.isnull().sum()
```

```python
missing=data[data.isnull().any(axis=1)]
missing
```

```python
data2=data.dropna(axis=0)
data2
```

```python
sal=data['SalStat']
```

```python
data2["SalStat"]=data2["SalStat"].map({' less than or equal to 50,000':0,' greater than 50,000':1})
print(data2["SalStat"])
```

```python
sal2=data2["SalStat"]
dfs=pd.concat([sal,sal2],axis=1)
dfs
```

```python
data2
```

```python
new_data=pd.get_dummies(data2,drop_first=True)
new_data
```

# FEATURE SELECTION METHODS


```python
column_list=list(new_data.columns)
print(column_list)
```

```python
features=list(set(column_list)-set(["SalStat"]))
print(features)
```

```python
y=new_data["SalStat"].values
print(y)
```

```python
x=new_data[features].values
x
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
kn_c=KNeighborsClassifier(n_neighbors=5)
kn_c.fit(train_x,train_y)
```

```python
predict=kn_c.predict(test_x)
```

```python
confusionmatrix=confusion_matrix(test_y,predict)
```

```python
confusionmatrix
```

```python
a=accuracy_score(test_y,predict)
a
```

```python
print("Misclassfied data: %d"%(test_y !=predict).sum())
```
